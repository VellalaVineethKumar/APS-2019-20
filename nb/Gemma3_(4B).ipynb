### Unsloth

`FastModel` supports loading nearly any model now! This includes Vision and Text models!

````python
from unsloth import FastModel
import torch

fourbit_models = [
    # 4bit dynamic quants for superior accuracy and low memory use
    "unsloth/gemma-3-1b-it-unsloth-bnb-4bit",
    "unsloth/gemma-3-4b-it-unsloth-bnb-4bit",
    "unsloth/gemma-3-12b-it-unsloth-bnb-4bit",
    "unsloth/gemma-3-27b-it-unsloth-bnb-4bit",

    # Other popular models!
    "unsloth/Llama-3.1-8B",
    "unsloth/Llama-3.2-3B",
    "unsloth/Llama-3.3-70B",
    "unsloth/mistral-7b-instruct-v0.3",
    "unsloth/Phi-4",
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastModel.from_pretrained(
    model_name = "unsloth/gemma-3-4b-it",
    max_seq_length = 2048, # Choose any for long context!
    load_in_4bit = True,  # 4 bit quantization to reduce memory
    load_in_8bit = False, # [NEW!] A bit more accurate, uses 2x memory
    full_finetuning = False, # [NEW!] We have full finetuning now!
    # token = "hf_...", # use one if using gated models
)
````

We now add LoRA adapters so we only need to update a small amount of parameters!

````python
model = FastModel.get_peft_model(
    model,
    finetune_vision_layers     = False, # Turn off for just text!
    finetune_language_layers   = True,  # Should leave on!
    finetune_attention_modules = True,  # Attention good for GRPO
    finetune_mlp_modules       = True,  # SHould leave on always!

    r = 8,           # Larger = higher accuracy, but might overfit
    lora_alpha = 8,  # Recommended alpha == r at least
    lora_dropout = 0,
    bias = "none",
    random_state = 3407,
)
````

### Data Prep
We now use the `Gemma-3` format for conversation style finetunes. We use [Maxime Labonne's FineTome-100k](https://huggingface.co/datasets/mlabonne/FineTome-100k) dataset in ShareGPT style. Gemma-3 renders multi turn conversations like below:

```
<bos><start_of_turn>user
Hello!<end_of_turn>
<start_of_turn>model
Hey there!<end_of_turn>
```

We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3, phi4, qwen2.5, gemma3` and more.

````python
from unsloth.chat_templates import get_chat_template
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma-3",
)
````

````python
# from datasets import load_dataset
# dataset = load_dataset("mlabonne/FineTome-100k", split = "train")
import pandas as pd
dataset = pd.read_csv('/content/5830b692.csv')
dataset.head()
````

````python
import pandas as pd
import json

def extract_all_json_attributes(df: pd.DataFrame) -> pd.DataFrame:
    """
    Extract ALL attributes from the REQ_JSON column.
    """
    extracted_data = []

    for idx, row in df.iterrows():
        record = {}

        # Keep original columns
        for col in df.columns:
            if col != 'REQ_JSON':
                record[col] = row[col]

        # Parse JSON if available
        if pd.notna(row['REQ_JSON']) and row['REQ_JSON']:
            try:
                json_data = json.loads(row['REQ_JSON'])

                # Extract based on whether it's a person or organization
                if 'person' in json_data:
                    person = json_data['person']
                    record['entity_type'] = 'person'

                    # Person basic attributes
                    record['firstName'] = person.get('firstName', '')
                    record['lastName'] = person.get('lastName', '')
                    record['person_country'] = person.get('country', '')

                    # Professional designation (YOU WERE MISSING THIS)
                    if 'professionalDesignation' in person:
                        if isinstance(person['professionalDesignation'], str):
                            record['professional_designation'] = person['professionalDesignation']
                        elif isinstance(person['professionalDesignation'], list) and person['professionalDesignation']:
                            record['professional_designation'] = person['professionalDesignation'][0].get('designation', '')

                    # Action details
                    if 'action' in person:
                        action = person['action']
                        record['action_type'] = action.get('type', '')
                        record['endpoint_id'] = action.get('endpointId', '')
                        record['endpoint_dcr_id'] = action.get('endpointDCRId', '')
                        record['requested_by'] = action.get('requestSubmittedBy', '')

                        if 'key' in action:
                            record['party_id'] = action['key'].get('partyId', '')

                    # Alternate identifier
                    if 'alternateIdentifier' in person:
                        identifiers = person['alternateIdentifier']
                        if isinstance(identifiers, list) and identifiers:
                            record['alternate_identifier'] = identifiers[0].get('alternateIdentifier', '')
                            record['alternate_identifier_type'] = identifiers[0].get('alternateIdentifierType', '')

                    # Electronic address
                    if 'electronicAddress' in person:
                        emails = person['electronicAddress']
                        if isinstance(emails, list) and emails:
                            record['email'] = emails[0].get('electronicAddress', '')
                            record['electronic_address_type'] = emails[0].get('electronicAddressType', '')  # YOU WERE MISSING THIS

                    # Phone communication
                    if 'phoneCommunication' in person:
                        phones = person['phoneCommunication']
                        if isinstance(phones, list) and phones:
                            record['phone'] = phones[0].get('phoneNumber', '')
                            record['phone_type'] = phones[0].get('phoneType', '')  # YOU WERE MISSING THIS

                    # Address
                    if 'postalAddress' in person and person['postalAddress']:
                        address = person['postalAddress'][0]
                        record['address_line1'] = address.get('addressLine1', '')
                        record['address_line2'] = address.get('addressLine2', '')
                        record['city'] = address.get('city', '')
                        record['state'] = address.get('state', '')
                        record['address_country'] = address.get('country', '')
                        record['postal_code'] = address.get('postalCode', '')

                        # Address action type (if needed)
                        if 'action' in address:
                            record['address_action_type'] = address['action'].get('type', '')

                elif 'organization' in json_data:
                    org = json_data['organization']
                    record['entity_type'] = 'organization'

                    # Organization attributes
                    record['org_name'] = org.get('name', '')
                    record['org_country'] = org.get('country', '')

                    # Action details
                    if 'action' in org:
                        action = org['action']
                        record['action_type'] = action.get('type', '')
                        record['endpoint_id'] = action.get('endpointId', '')
                        record['endpoint_dcr_id'] = action.get('endpointDCRId', '')
                        record['requested_by'] = action.get('requestSubmittedBy', '')

                        if 'key' in action:
                            record['party_id'] = action['key'].get('partyId', '')

                    # Address
                    if 'postalAddress' in org and org['postalAddress']:
                        address = org['postalAddress'][0]
                        record['address_line1'] = address.get('addressLine1', '')
                        record['address_line2'] = address.get('addressLine2', '')
                        record['city'] = address.get('city', '')
                        record['state'] = address.get('state', '')
                        record['address_country'] = address.get('country', '')
                        record['postal_code'] = address.get('postalCode', '')

                        # Address action type (if needed)
                        if 'action' in address:
                            record['address_action_type'] = address['action'].get('type', '')

            except json.JSONDecodeError as e:
                record['json_parse_error'] = str(e)
            except Exception as e:
                record['extraction_error'] = str(e)

        extracted_data.append(record)

    return pd.DataFrame(extracted_data)
````

We now use `standardize_data_formats` to try converting datasets to the correct format for finetuning purposes!

````python
def create_conversations(row):
    # Exclude target columns from input
    input_attributes = {
        col: row[col] for col in row.index
        if col not in ['DS Review', 'Comment', 'Unnamed: 11'] # Also exclude Unnamed: 11 if it's not relevant
    }

    # Format input attributes into a string
    input_text = "Given the following attributes:\n"
    for key, value in input_attributes.items():
        input_text += f"{key}: {value}\n"

    # Format the target output
    target_text = f"DS Review: {row['DS Review']}\nComment: {row['Comment']}"

    # Create the conversation structure
    conversation = [
        {"role": "user", "content": [{"type": "text", "text": input_text}]},
        {"role": "assistant", "content": [{"type": "text", "text": target_text}]}
    ]

    return {"conversations": conversation, "source": "custom_dataset", "score": 1.0}

# Apply the function to the dataset to create the new format
dataset = dataset.apply(create_conversations, axis=1)
````

We now have to apply the chat template for `Gemma-3` onto the conversations, and save it to `text`. We remove the `<bos>` token using removeprefix(`'<bos>'`) since we're finetuning. The Processor will add this token before training and the model expects only one.

````python
def formatting_prompts_func(examples):
   convos = examples["conversations"]
   texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False).removeprefix('<bos>') for convo in convos]
   return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)
````

<a name="Train"></a>
### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

````python
from trl import SFTTrainer, SFTConfig
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None, # Can set up evaluation!
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4, # Use GA to mimic batch size!
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 30,
        learning_rate = 2e-4, # Reduce to 2e-5 for long training runs
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "none", # Use this for WandB etc
    ),
)
````

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs. This helps increase accuracy of finetunes!

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<start_of_turn>user\n",
    response_part = "<start_of_turn>model\n",
)
````

### Inference
Let's run the model via Unsloth native inference! According to the `Gemma-3` team, the recommended settings for inference are `temperature = 1.0, top_p = 0.95, top_k = 64`

````python
from unsloth.chat_templates import get_chat_template
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma-3",
)
messages = [{
    "role": "user",
    "content": [{
        "type" : "text",
        "text" : "Continue the sequence: 1, 1, 2, 3, 5, 8,",
    }]
}]
text = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True, # Must add for generation
)
outputs = model.generate(
    **tokenizer([text], return_tensors = "pt").to("cuda"),
    max_new_tokens = 64, # Increase for longer outputs!
    # Recommended Gemma-3 settings!
    temperature = 1.0, top_p = 0.95, top_k = 64,
)
tokenizer.batch_decode(outputs)
````

 You can also use a `TextStreamer` for continuous inference - so you can see the generation token by token, instead of waiting the whole time!

````python
messages = [{
    "role": "user",
    "content": [{"type" : "text", "text" : "Why is the sky blue?",}]
}]
text = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True, # Must add for generation
)

from transformers import TextStreamer
_ = model.generate(
    **tokenizer([text], return_tensors = "pt").to("cuda"),
    max_new_tokens = 64, # Increase for longer outputs!
    # Recommended Gemma-3 settings!
    temperature = 1.0, top_p = 0.95, top_k = 64,
    streamer = TextStreamer(tokenizer, skip_prompt = True),
)
````
